```python
import pandas as pd
from matplotlib import pyplot as plt

base_path = "./analysis/result/"
ch_path = base_path + "ch_manufac/growth_accounting.csv"
jn_path = base_path + "jn_manufac/growth_accounting.csv"
sg_path = base_path + "sg_manufac/growth_accounting_by_real_gos.csv"
tw_path = base_path + "tw_manufac/growth_accounting.csv"

ch = pd.read_csv(ch_path, index_col=0)
jn = pd.read_csv(jn_path, index_col=0)
sg = pd.read_csv(sg_path, index_col=0)
tw = pd.read_csv(tw_path, index_col=0)

tfp_trend = pd.DataFrame({
  "China": ch["Contribution of TFP(Trend)"],
  "Japan": jn["Contribution of TFP(Trend)"],
  "Singapore": sg["TFP contribution (Trend)"],
  "Taiwan": tw["TFP contribution (Trend)"],
})
```

```python

plt.plot(tfp_trend, label=tfp_trend.columns)
plt.title("TFP trend comparison")
plt.xlabel("year")
# plt.xticks(rotation=45)
plt.ylabel("contribution rate")
plt.legend()
plt.show()
```

```python
extraction = ["Japan", "Singapore", "Taiwan"]
plt.plot(tfp_trend[extraction], label=extraction)
plt.title("TFP trend comparison")
plt.xlabel("year")
# plt.xticks(rotation=45)
plt.ylabel("contribution rate")
plt.legend()
plt.show()
```
